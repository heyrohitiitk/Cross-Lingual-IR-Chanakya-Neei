```python
import pyiwn 
iwn = pyiwn.IndoWordNet(lang=pyiwn.Language.HINDI) 
```

```python
file = open("final_words.csv",encoding="utf-8").read()
```

```python
words = file.split(",")
len(words)
```

```python
import pandas as pd
base_words = pd.read_pickle('Base_words_list.pkl')
```

```python
final_words = pd.read_pickle("final_words.pkl")
```

```python
final_words = set(final_words.values())
```

```python

final_words = final_words.union(base_words)
```

```python
len(final_words)
```

```python
synonyms = {}
for word in final_words:
  try:
    synsets = iwn.synsets(word)
    for synset in synsets:
      lemmas = synset.lemma_names()
      for lemma in lemmas:
        if lemma in synonyms.keys():
          synonyms[lemma].append(word)
        else:
          synonyms[lemma]=[word]
  except:
    pass
```

```python
synonyms
```

```python
import pickle
with open('synonyms.pkl', 'wb') as file:
    pickle.dump(synonyms, file)
```
